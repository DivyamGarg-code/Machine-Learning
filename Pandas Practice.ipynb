
# Terrorism Data

```python
d=pd.read_csv('terrorismData.csv')
df=d.copy()
df
```

```python
# which city in Jammu and Kashmir has the maximum attacks
a=df[df.State=='Jammu and Kashmir']
l=list(a.City)
d={}
for i in l:
    if i not in d.keys():
        d[i]=1;
    else:
        d[i]+=1;
d_v=list(d.values())
d_k=list(d.keys())
max_attacks=max(d_v)
dan_place=d_k[d_v.index(max_attacks)]
print(f"most dangerous place--> '{dan_place}' with the maximum number of attacks --> {max_attacks}")

```

```python
# country with the maximum number of terror attacks
l=list(df.Country)
d={}
for i in l:
    if i not in d.keys():
        d[i]=1;
    else:
        d[i]+=1;
#print(d)
d_v=list(d.values()) # count of the number of attacks
d_k=list(d.keys()) # in which country
max_attacks=max(d_v)
dan_country=d_k[d_v.index(max_attacks)] # dangerous country
print(f"most dangerous country--> '{dan_country}' with the maximum number of attacks --> {max_attacks}")
```

```python
# country with the maximum number of terror attacks occured in which year?
# Let's consider the data for that particular country
b=df[df.Country==dan_country]
l=list(b.Year)
d={} # key --> Year value--> count
for i in l:
    if i not in d.keys():
        d[i]=1; 
    else:
        d[i]+=1;
print(d)
d_v=list(d.values()) # count of the number of attacks
d_k=list(d.keys()) # in which year
max_attacks=max(d_v)
year=d_k[d_v.index(max_attacks)] # dangerous country
print(f"most dangerous country--> '{dan_country}' with the maximum number of terror attacks --> {max_attacks} in year {year}")
```

```python
# Find deadliest attack(maximum number of people killed), which Country, Terrorist_Group ?
```

```python
df.isnull().sum()
```

```python
a=df.dropna() # This will remove the rows with NaN entries because here it is the best consideration to ignore data which is not known
a.reset_index(drop=True,inplace=True)
l=list(a.Country)
d={} # key --> Country || value--> count of no. of people killed
for i in l:
    if i not in d.keys():
        c=a[a.Country==i] # this will create the data for that country
        cnt=c.Killed.count()
        d[i]=cnt; 
print(d)
d_v=list(d.values()) # count of the number of people killed
d_k=list(d.keys()) # in which country
max_killed=max(d_v)
country=d_k[d_v.index(max_killed)] # dangerous country
print(f" country--> '{country}' with the maximum number of people killed --> {max_killed}")
```

```python
b=a[a.Country==country]
l=list(b.AttackType)
d={} # key --> AttackType || value--> count of this AttackType
for i in l:
    if i not in d.keys():
        d[i]=1
    else:
        d[i]+=1
print(d)
d_v=list(d.values()) # count of the number of people killed
d_k=list(d.keys()) # in which country
max_times=max(d_v)
Attacktype=d_k[d_v.index(max_times)] # dangerous country
print(f" country--> '{country}' with the maximum attack  by --> {Attacktype}  {max_times}times")
c=b[b.AttackType==Attacktype]
cnt=c.Killed.sum()
print(f"and in that total number of people killed are {cnt}")
```

```python
import numpy as nm
import csv
import time
```

```python
# Finding the sum of total number of people (killed+wounded)  for country India and this is working fine for 'year2017.csv' file
t1=time.time() 
file_obj=open('year2017.csv','r')
file_data=csv.DictReader(file_obj,skipinitialspace=True)
killed=[]
wounded=[]
country=[]
for row in file_data:
    killed.append(row['Killed'])
    wounded.append(row['Wounded'])
    country.append(row['Country'])
killed=nm.array(killed)
wounded=nm.array(wounded)

killed[killed=='']='0' # this will replace the missing values/empty string with '0'
wounded[wounded=='']='0' # this will replace the missing values/empty string with '0'

killed=nm.array(killed,dtype=float)
wounded=nm.array(wounded,dtype=float)
country=nm.array(country)

bool=country=='India' # this will make an array containing true or false depending on the condition

k_w=killed+wounded 

print(sum(k_w)) # sum of killed and wounded people for all the countries

print(sum(k_w[bool])) # sum of killed and wounded people for country=India

t2=time.time()
print(f"time taken to be executed is {(t2-t1)}s")
```

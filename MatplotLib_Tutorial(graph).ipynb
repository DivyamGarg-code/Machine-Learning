```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=[1,2,3]
y=[2,4,6]
plt.scatter(x,y) # scatter/show the POINTS on the graph
plt.show() # after performing operations you can call this function to show the changes being made
```

```python
# If you want to decide the color of your line you can pass a string as a parameter ,It's 1st character would decide the color and the 2nd character would
# decide its form(points(with shape-circle,plus,triangle),line(dashed or complete) and many more)
plt.plot(x,y,"ro")
plt.plot(x,y,"g-.")
```

```python
# plotting x^3
x=np.arange(0,5,0.1)
y=x**3
plt.plot(x,y)
```

```python
x=np.arange(0,5,0.1)
y=x**3
plt.plot(x,y,color="blue",marker='*')
```

```python
x=np.arange(0,5,0.1)
y=x**3
y1=x**2
plt.plot(x,y,color="blue",linewidth=5) # You can also set the linewidth
# To do the labeling on x and y axis
plt.xlabel("x ")
plt.ylabel("x^3 ")
plt.title("Matplotlib Demo",color="red")
```

```python
x=np.arange(0,5,0.1)
y=x**3
y1=x**2
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.plot(x,y,color="black",linewidth=2,label="x**3")
plt.plot(x,y1,color="red",linewidth=2,label="x**2")
plt.legend() # this will show the difference b/w 2 diifer lines y(x) based on their color 
plt.axis([0,2,0,2]) # this will take 4 arguments range ox x and y to be shown plt.axis(x1,x2,y1,y2)
plt.grid(True) # will show the grid lines
# Adding the random text as some place
plt.text(1.1,1.1,"Point of Intersection",fontsize=15)
```

BUBBLE GRAPHS

```python
year=[2012,2013,2014,2015,2016,2017]
salary=[12,13,14,17,19,20]
population=[100,120,180,250,300,370]
plt.xlabel("Year",fontsize=15)
plt.ylabel("Salary",fontsize=15)
plt.scatter(year,salary,s=100) # s -> size ie all the the bubbles are at the same size of s salary vs year graph
```

```python
# But suppose if you want to get the idea about the size of the workers working for the salary in that particular year 
plt.scatter(year,salary,s=population,c='r',alpha=0.5) # c -> color alpha-> opacity more less it is more transparent it is lying b/w (0,1)
```

```python
# If you want to pass different/array of colors to the different points
c=np.arange(len(year))
plt.title("Bubble Chart")
plt.scatter(year,salary,s=population,c=c,alpha=1) 
```

```python
plt.xlabel("Year",fontsize=15)
plt.ylabel("Salary",fontsize=15)
plt.scatter(year,salary,s=population,c=c,alpha=1,edgecolors="black",marker="^") # You can also the edge colors
# Showing the population at the points at that distancex,y is the location at which you want to write text
for i in range(0,len(year)):                                       
    plt.text(year[i]+0.1,salary[i]+0.1,population[i]) # plt.text(x,y,text) 
plt.legend()
```

PIE GRAPHS

```python
# Splitting the graph based on the sizes
sizes=[3,4,6,2]
labels=['A','B','C','D']
colors=['blue','purple','red','pink']
# To have the percentage distribution among these 4 we have autopct
plt.pie(sizes,labels=labels,colors=colors,autopct="%.1f%%")
plt.show()
```

```python
# If you want to highlight some class we have explode property
explode=[0.1,0,0,0] # set for all the values
plt.pie(sizes,labels=labels,colors=colors,autopct="%.1f%%",explode=explode)
plt.show()
```

```python
# By default it moves in counterclockwise direction so to change that put counterclock=false
plt.pie(sizes,labels=labels,colors=colors,autopct="%.2f%%",explode=explode,counterclock=False,startangle=90)
plt.show()
```

Histogram

```python
# To measure the frequency,range we use the histogram
a=[1,2,3,4,1,3,4,2,1,1]
plt.hist(a)
```

```python
b=[1,2,1,3,5,4,5,5,4,3,3,10,7,6,7,8,8,20,22,11,14]
plt.hist(b)
plt.show() # This somehow shows the combined frequency of some numbers 
```

```python
b=[1,2,1,3,5,4,5,5,4,3,3,10,7,6,7,8,8,20,22,11,14]
xt=np.arange(23)
plt.xticks(xt) # will show the numbers on x axis
plt.hist(b,edgecolor='Black')
# plt.show()
# By default it create 10 towers/bins like from 1-3.1 , 3.1-5.2 , 5.2-7.3 and so on..  and the size of each bin is calculated by (max_no - min_no)/bins(10)
# Here it is (22-1)/10 ie 2.1 and you can see each bin with size 2.1 and the combined frequency of numbers lying in these bins like 1-3.1 , 3.1-5.2 , 5.2-7.3 
# are calculated and histogram is drawn
# Here you can relate 10 bins as 10 strips generated with the width as calculated above but you can also set the no. of bins 
```

```python
plt.hist(b,bins=21,edgecolor='black')
plt.xticks(xt)
plt.show()
# [a,b] here in the bins like [1,2) [2,3) a is included and b is not but in the last bin both a and b are included ie [21,22]
```

```python
plt.hist(b,bins=[1,3,6,10,15,20],edgecolor='Black') # You can also set the bins range flexiblly
# This will create bins ranging from [1,3) [3,6) [6,10) [10,15) [15,20] and adding the frequencies of numbers lying in these ranges
```

```python
plt.hist(b,bins=[1,3,6,10,15,20],edgecolor='Black')
plt.xticks([1,3,6,10,15,20])
plt.yticks(np.arange(11))
plt.show()
# plt.axis([0,25,0,10]) You can set range of axis also
```

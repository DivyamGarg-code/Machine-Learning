# Handling NaN / Missing values

```python
# Replacing the Nan values with something
df=data.copy()
df.columns=['country','age','income','purchased']
# df.dropna() will simply drop the nan values
df.age.fillna(df.age.mean(),inplace=True)
df.income.fillna(df.income.mean(),inplace=True)
df
```
